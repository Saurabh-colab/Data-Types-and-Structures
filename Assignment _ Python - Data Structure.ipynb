##Data Types and Structures Questions - Theoritical




1. **What are data structures, and why are they important?**
- Data structures are like different ways to organize and store information in a computer. They're important because they help us find, add, or remove that information quickly and easily, making programs work better.

2. **Explain the difference between mutable and immutable data types with examples.**
- Mutable (Changeable): You can change the content of these data types after you've created them.

- Immutable (Unchangeable): You cannot change the content of these data types once they are created.

```python
#Example -
#Mutable -

my_list = [1, 2, 3]
```

```python
my_list[0] = 10  # List changed
```

```python
my_list
```

```python
#Immutable
my_string = "hello"
```

```python
my_string[0] = "H"  # Error! Strings can't be changed
```

3. **What are the main differences between lists and tuples in Python**
- Mutability:

List is mutable (can be changed).

Tuple is immutable (cannot be changed).

- Syntax:

List: [1, 2, 3]

Tuple: (1, 2, 3)

- Performance:

Tuples are faster than lists.

- Use case:

Use lists when data can change.

Use tuples for fixed data.

4. **Describe how dictionaries store data?**

- Dictionaries store data in key-value pairs.

Each item has a key and a value.

You use the key to find the value quickly.

Example:

student = {"name": "saurabh", "age": 20}
"name" and "age" are keys

"saurabh" and 20 are values

You can get data like: student["name"] → "saurabh"

**5.Why might you use a set instead of a list in Python**

 A set is best when:

- we need only unique items (no duplicates).

- we need to quickly check if an item is present.

- The order of items doesn't matter.

**6. What is a string in Python, and how is it different from a list?**

- A string is a sequence of characters, while a list is a collection of items that can be of different types.
Strings are immutable, but lists are mutable.

**7. How do tuples ensure data integrity in Python**

- Tuples ensure data integrity by being immutable, meaning their values cannot be changed after creation.

**8. What is a hash table, and how does it relate to dictionaries in Python**

- A hash table is a data structure that stores data using key-value pairs and allows fast access using keys.
In Python, dictionaries are built using hash tables.

**9.  Can lists contain different data types in Python?**

Yes, lists in Python can contain different data types in the same list.

**10. Explain why strings are immutable in Python?**

- Strings are immutable in Python to make them faster and safer to use.

**11. What advantages do dictionaries offer over lists for certain tasks?**

- Dictionaries offer faster data lookup using keys and can store data with meaningful labels, unlike lists that use only index numbers.

**12. Describe a scenario where using a tuple would be preferable over a list?**

- Using a tuple is better when we want to store fixed data that should not change, like coordinates (x, y) or a person's birthdate.

**13.  How do sets handle duplicate values in Python?**
- Sets automatically remove duplicate values and keep only unique items.

**14. How does the “in” keyword work differently for lists and dictionaries?**

- For lists, “in” checks if a value is in the list; for dictionaries, it checks if a key exists.

**15. Can you modify the elements of a tuple? Explain why or why not?**

- No, you cannot modify elements of a tuple because tuples are immutable, meaning their content cannot be changed after creation.


**16. What is a nested dictionary, and give an example of its use case?**

- A nested dictionary is a dictionary inside another dictionary.

Example use: Storing information about students, where each student has details like name and grades

students = {
  "student1": {"name": "Alice", "grade": 90},

  "student2": {"name": "Bob", "grade": 85}

}

**17.  Describe the time complexity of accessing elements in a dictionary.**

- Accessing elements in a dictionary is usually very fast, with an average time complexity of O(1) (constant time).

**18.  In what situations are lists preferred over dictionaries?**

- Lists are preferred when we need ordered data or want to store items without keys, especially if you need to access items by position.

**19. Why are dictionaries considered unordered, and how does that affect data retrieval?**

- Dictionaries are considered unordered because their items don’t have a fixed order, so you can’t rely on the order when retrieving data; instead, you use keys to access values directly.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
- Lists retrieve data by position (index), while dictionaries retrieve data by keys.
